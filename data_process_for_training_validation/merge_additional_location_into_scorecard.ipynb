```python
#describe each location with companies in side
import pandas as pd
import numpy as np
import os
import pygeohash as pgh
from math import *
from sklearn.metrics.pairwise import euclidean_distances
import random
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression as LR
from sklearn.metrics import precision_recall_curve, roc_curve, auc
from matplotlib import pylab
from sklearn.preprocessing import normalize
pjoin = os.path.join
```

```python
#data load
datapath = '/Users/yefeichen/Database/location_recommender_system/'
lfile = 'location_scorecard_190912.csv'
lfile_add = 'WeWork_buildings_from_wulei.csv'

loc_feat = pd.read_csv(pjoin(datapath,lfile))
col_list = list(loc_feat.columns)
loc_feat_ww = pd.read_csv(pjoin(datapath,lfile_add))
#column alignment
loc_feat_ww = loc_feat_ww[col_list]
#flag
loc_feat_ww['source'] = 'newest'
```

```python
loc_feat = loc_feat.merge(loc_feat_ww[['atlas_location_uuid','source']],on='atlas_location_uuid',how='left',suffixes=['','_right'])
loc_feat_remain = loc_feat[loc_feat['source']!='newest']
#col alignment
loc_feat_remain = loc_feat_remain[col_list]
# loc_feat_remain.head()
loc_feat_all = pd.concat([loc_feat_remain,loc_feat_ww],axis=0,sort='False')
# loc_feat_all.head()
```

```python
#check and save
assert(len(loc_feat_all) == len(loc_feat_all.groupby('atlas_location_uuid').first()))
loc_feat_all.to_csv(pjoin(datapath,'location_scorecard_191113.csv'))
```
